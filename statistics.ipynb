```python
import pandas as pd
import numpy as np
import math
```

```python
class Statistics:
    def __init__(self, data) -> None:
        self.data = pd.DataFrame({"x" : data})
        
    def mean(self):
        return self.data.mean()
    
    def weighted_mean(self, weighting_data):
        return (self.data.values * weighting_data.values).sum() / weighting_data.sum()
    
stats = Statistics([1,2,3,4])
stats.mean()
stats.weighted_mean(pd.DataFrame({"w":[5,4,3,2]}))
```
